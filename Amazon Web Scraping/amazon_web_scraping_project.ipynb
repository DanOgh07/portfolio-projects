# Connecting to Webpage(s)

```python
URL = "https://www.amazon.com/dp/B09PRK9C47"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/148.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
}

page = requests.get(URL, headers=headers)

soup = BeautifulSoup(page.content, "html.parser")
soup = BeautifulSoup(soup.prettify())

title = soup.find(id='productTitle').get_text(strip=True)

'''
symbol = soup.find('span', {'class': 'a-price-symbol'}).get_text()
whole = soup.find('span', {'class': 'a-price-whole'}).get_text()
fraction = soup.find('span', {'class': 'a-price-fraction'}).get_text()

price = f"{symbol}{whole}{fraction}"
'''


price = soup.find('span', {'class': 'a-price'}).get_text(strip=True)


print(title)
print(price) 
```

```python
# Removing the currency symbol (in NGN, not $) from the price string

price = price[3:]
print(price)
```
