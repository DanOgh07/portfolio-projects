```python
def flag_emoji_to_iso(emoji_str):
    """'🇨🇳' → 'cn'  (works for any flag emoji)"""
    ri_chars = [c for c in emoji_str if 0x1F1E6 <= ord(c) <= 0x1F1FF]
    if len(ri_chars) == 2:
        return ''.join(chr(ord(c) - 0x1F1E6 + ord('a')) for c in ri_chars)
    return None

country_iso = {
    country: flag_emoji_to_iso(emoji)
    for country, emoji in country_flag_emoji.items()
    if flag_emoji_to_iso(emoji)
}
print(f"ISO codes mapped: {len(country_iso)}")
```

```python
flag_cache = {}
failed = []

for country, iso in country_iso.items():
    try:
        url = f'https://flagcdn.com/w40/{iso}.png'
        resp = requests.get(url, timeout=5)
        resp.raise_for_status()
        img = Image.open(BytesIO(resp.content)).convert('RGBA')
        flag_cache[country] = np.array(img)
    except Exception as e:
        failed.append(f'{country} ({iso}): {e}')

print(f"Downloaded: {len(flag_cache)}  |  Failed: {len(failed)}")
for f in failed:
    print(' ', f)
```

```python
_orig_anim_func = _mc._BarChartRace.anim_func

def _anim_with_flags(self, i):
    # Let BCR draw the bars as normal
    _orig_anim_func(self, i)

    # Remove flag images placed in the previous frame
    for artist in list(self.ax.get_children()):
        if getattr(artist, '_is_flag', False):
            artist.remove()

    # Determine which countries are visible and their y-positions this frame
    bar_location = self.df_ranks.iloc[i].values
    top_filt = (bar_location > 0) & (bar_location < self.n_bars + 1)
    cols      = self.df_values.columns[top_filt]
    y_positions = bar_location[top_filt]

    # Blended transform: x in axes-fraction, y in data coordinates
    # So xy=(0, y) means "left edge of the plot, at this bar's y-position"
    transform = blended_transform_factory(self.ax.transAxes, self.ax.transData)

    for country, y_pos in zip(cols, y_positions):
        if country not in flag_cache:
            continue
        imagebox = OffsetImage(flag_cache[country], zoom=0.50)
        ab = AnnotationBbox(
            imagebox,
            xy=(0, y_pos),
            xycoords=transform,
            box_alignment=(0.0, 0.5),   # left edge of image at left edge of plot
            pad=0,
            frameon=False,
        )
        ab._is_flag = True              # tag so we can remove it next frame
        self.ax.add_artist(ab)

_mc._BarChartRace.anim_func = _anim_with_flags
print("BCR patched — flags will be injected per frame.")
```
